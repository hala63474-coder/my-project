```python
import os
import cv2
import tensorflow as tf
```

```python
fake_videos_path = '/content/drive/MyDrive/dataset/fake'
real_videos_path = '/content/drive/MyDrive/dataset/real'
output_path = '/content/drive/MyDrive/frames'

fake_folder = os.path.join(output_path, 'fake')
real_folder = os.path.join(output_path, 'real')

os.makedirs(fake_folder, exist_ok=True)
os.makedirs(real_folder, exist_ok=True)
```

```python
def extract_frames(video_path, save_folder, label, max_frames=50):
    """
    video_path: مسار الفيديو
    save_folder: مجلد الحفظ
    label: 'fake' أو 'real'
    max_frames: عدد أقصى للفريمات من كل فيديو
    """
    cap = cv2.VideoCapture(video_path)
    count = 0

    while cap.isOpened() and count < max_frames:
        ret, frame = cap.read()
        if not ret:
            break

        filename = f"{label}_{os.path.basename(video_path)}_{count}.jpg"
        filepath = os.path.join(save_folder, filename)

        if not os.path.exists(filepath):  # لو الصورة موجودة مسبقاً لا ننسخ
            cv2.imwrite(filepath, frame)
        count += 1

    cap.release()
```

```python
import os

if not os.path.exists(output_path) or len(os.listdir(output_path)) == 0:

    print("Extracting frames...")

    # Fake videos
    for video in os.listdir(fake_videos_path)[:100]:
        extract_frames(os.path.join(fake_videos_path, video), fake_folder, 'fake')

    # Real videos
    for video in os.listdir(real_videos_path)[:100]:
        extract_frames(os.path.join(real_videos_path, video), real_folder, 'real')

    print("Finished extracting frames")

else:
    print("Frames already exist, skipping extraction")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

batch_size = 32
img_height = 224
img_width = 224

train_data = datagen.flow_from_directory(
    output_path,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',
    subset='training'
)

val_data = datagen.flow_from_directory(
    output_path,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',
    subset='validation'
)

print("✅ Data ready for training")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),

    Dense(128, activation='relu'),
    Dropout(0.5),

    Dense(1, activation='sigmoid')  # binary: fake / real
])

model.summary()
```

```python
from tensorflow.keras.models import load_model

model_path = '/content/drive/MyDrive/deepfake_model.h5'

model = load_model(model_path, compile=False)

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("🔥 الموديل جاهز")
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

# 📌 مسار الصورة
img_path = '/content/drive/MyDrive/frames/real/real_10__walking_down_street_outside_angry.mp4_1.jpg'

# 📌 تحميل الصورة
img = image.load_img(img_path, target_size=(224,224))

# 📌 تحويلها لـ array
img_array = image.img_to_array(img)

# 📌 تجهيزها للموديل
img_array = np.expand_dims(img_array, axis=0) / 255.0

# 📌 التوقع
prediction = model.predict(img_array)

# 📌 النتيجة + النسبة
confidence = prediction[0][0]

if confidence > 0.5:
    print(f"Real ✅ بنسبة {confidence*100:.2f}%")
else:
    print(f"Fake ❌ بنسبة {(1-confidence)*100:.2f}%")
```
