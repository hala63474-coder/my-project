```python
def extract_frames(video_path, save_folder, label, max_frames=50):
    """
    video_path: مسار الفيديو
    save_folder: مجلد الحفظ
    label: 'fake' أو 'real'
    max_frames: عدد أقصى للفريمات من كل فيديو
    """
    cap = cv2.VideoCapture(video_path)
    count = 0

    while cap.isOpened() and count < max_frames:
        ret, frame = cap.read()
        if not ret:
            break

        filename = f"{label}_{os.path.basename(video_path)}_{count}.jpg"
        filepath = os.path.join(save_folder, filename)

        if not os.path.exists(filepath):  # لو الصورة موجودة مسبقاً لا ننسخ
            cv2.imwrite(filepath, frame)
        count += 1

    cap.release()
```

```python
import os

output_path = '/content/drive/MyDrive/deepfake_project/frames'

fake_folder = os.path.join(output_path, 'fake')
real_folder = os.path.join(output_path, 'real')

# ✅ عد الصور فقط
def count_frames(folder):
    if not os.path.exists(folder):
        return 0
    return len([
        f for f in os.listdir(folder)
        if f.lower().endswith(('.jpg', '.png', '.jpeg'))
    ])

# ✅ شرط الاستخراج
if not os.path.exists(fake_folder) or not os.path.exists(real_folder) or \
   count_frames(fake_folder) == 0 or count_frames(real_folder) == 0:

    print("Extracting frames...")

    for video in os.listdir(fake_videos_path)[:100]:
        extract_frames(os.path.join(fake_videos_path, video), fake_folder, 'fake')

    for video in os.listdir(real_videos_path)[:100]:
        extract_frames(os.path.join(real_videos_path, video), real_folder, 'real')

    print("Finished extracting frames")

else:
    print("Frames already exist ✔, skipping extraction")

# 📊 العد النهائي
fake_count = count_frames(fake_folder)
real_count = count_frames(real_folder)
total = fake_count + real_count

print(f"\nFake frames : {fake_count}")
print(f"Real frames : {real_count}")
print(f"Total       : {total}")
```

```python
import os
import cv2

fake_videos_path2 = '/content/drive/MyDrive/deepfake_project/dataset2/Fake'
real_videos_path2 = '/content/drive/MyDrive/deepfake_project/dataset2/Real'

fake_folder2 = '/content/drive/MyDrive/deepfake_project/frames2/Fake'
real_folder2 = '/content/drive/MyDrive/deepfake_project/frames2/Real'

os.makedirs(fake_folder2, exist_ok=True)
os.makedirs(real_folder2, exist_ok=True)

def count_frames(folder):
    if not os.path.exists(folder):
        return 0
    return len([f for f in os.listdir(folder) if f.lower().endswith(('.jpg', '.png', '.jpeg'))])

def extract_frames2(video_path, save_folder, label, max_frames=30):
    cap = cv2.VideoCapture(video_path)
    count = 0
    while cap.isOpened() and count < max_frames:
        ret, frame = cap.read()
        if not ret:
            break
        filename = f"{label}_{os.path.basename(video_path)}_{count}.jpg"
        filepath = os.path.join(save_folder, filename)
        if not os.path.exists(filepath):
            cv2.imwrite(filepath, frame)
        count += 1
    cap.release()

# ✅ شرط: إذا الفريمات موجودة تخطى الاستخراج
if count_frames(fake_folder2) > 0 and count_frames(real_folder2) > 0:
    print("Frames already exist ✔, skipping extraction")
else:
    print("Extracting frames from dataset2...")
    for video in os.listdir(fake_videos_path2):
        extract_frames2(os.path.join(fake_videos_path2, video), fake_folder2, 'fake')
    for video in os.listdir(real_videos_path2):
        extract_frames2(os.path.join(real_videos_path2, video), real_folder2, 'real')
    print("✅ Finished!")

print(f"Fake frames: {count_frames(fake_folder2)}")
print(f"Real frames: {count_frames(real_folder2)}")
print(f"Total: {count_frames(fake_folder2) + count_frames(real_folder2)}")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# مسار يجمع القديمة والجديدة
# بنعمل داتا جنريتور من كلا المجلدين

from tensorflow.keras.utils import image_dataset_from_directory
import tensorflow as tf

# جمع كل الفريمات
def get_all_files(folders):
    files = []
    labels = []
    for folder, label in folders:
        for f in os.listdir(folder):
            if f.lower().endswith(('.jpg', '.png', '.jpeg')):
                files.append(os.path.join(folder, f))
                labels.append(label)
    return files, labels

import os
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import pandas as pd

# كل الفريمات
fake_old = '/content/drive/MyDrive/deepfake_project/frames/fake'
real_old = '/content/drive/MyDrive/deepfake_project/frames/real'
fake_new = '/content/drive/MyDrive/deepfake_project/frames2/Fake'
real_new = '/content/drive/MyDrive/deepfake_project/frames2/Real'

# جمع كل الملفات
all_files = []
all_labels = []

for f in os.listdir(fake_old):
    if f.lower().endswith(('.jpg', '.png', '.jpeg')):
        all_files.append(os.path.join(fake_old, f))
        all_labels.append('fake')

for f in os.listdir(real_old):
    if f.lower().endswith(('.jpg', '.png', '.jpeg')):
        all_files.append(os.path.join(real_old, f))
        all_labels.append('real')

for f in os.listdir(fake_new):
    if f.lower().endswith(('.jpg', '.png', '.jpeg')):
        all_files.append(os.path.join(fake_new, f))
        all_labels.append('fake')

for f in os.listdir(real_new):
    if f.lower().endswith(('.jpg', '.png', '.jpeg')):
        all_files.append(os.path.join(real_new, f))
        all_labels.append('real')

print(f"Total files: {len(all_files)}")
print(f"Fake: {all_labels.count('fake')}")
print(f"Real: {all_labels.count('real')}")

# تقسيم train/val
df = pd.DataFrame({'filename': all_files, 'label': all_labels})
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['label'])

print(f"\nTrain: {len(train_df)}")
print(f"Val: {len(val_df)}")

# داتا جنريتور
datagen = ImageDataGenerator(rescale=1./255)

train_data = datagen.flow_from_dataframe(
    train_df,
    x_col='filename',
    y_col='label',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

val_data = datagen.flow_from_dataframe(
    val_df,
    x_col='filename',
    y_col='label',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

print("✅ Data ready!")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout,GlobalAveragePooling2D

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    GlobalAveragePooling2D(),

    Dense(128, activation='relu'),
    Dropout(0.5),

    Dense(1, activation='sigmoid')  # binary: fake / real
])

model.summary()
```

```python
from tensorflow.keras.models import load_model

model_path = '/content/drive/MyDrive/deepfake_model.h5'

#model = load_model(model_path, compile=False)

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("🔥 الموديل جاهز")
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    train_data,
    validation_data=val_data,|
    epochs=15,
    callbacks=[early_stop]
)
```

```python
loss, acc = model.evaluate(val_data)
print(f"Model Accuracy:{acc*100:.2f}%")
```
